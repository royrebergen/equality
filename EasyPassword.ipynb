```python
import pandas as pd
import re
```

```python
def f(x):
    return len(x['woord_schoon'])

def r(x):
    x = re.sub(r'\W+', '', x['woord'])
    return x

def sideLen(x):
    x = x['woord_schoon'].lower()
    #left
    l_count = 0
    #right
    r_count = 0
    
    for y in x:
        if y in keys['l_keys']:
            l_count += 1
        elif y in keys['r_keys']:
            r_count += 1
            
    return pd.Series({'l_count':l_count,
                      'r_count':r_count,
                      'l_per':(round((l_count/len(x))*100)),
                      'r_per':(round((r_count/len(x))*100))})

```

```python
wordFile = open('data/woordenlijst.txt', 'r')
wordList = [line.split('\n') for line in wordFile.readlines()]
```

```python
df = pd.read_csv('data/woordenlijst.txt', sep="\n", header=None)
df.columns = ["woord"]
df['woord_schoon'] = df.apply(r, axis=1)
```

```python
#left-side keys
keys = { 'l_keys':['q','w','e','r','t','a','s','d','f','g','z','x','c','v'],
         'r_keys':['y','u','i','o','p','h','j','k','l','b','n','m']}
```

```python
df['lengte'] = df.apply(f, axis=1)
```

```python
df[['l_count','r_count','l_per','r_per']] = df.apply(sideLen,axis=1)
```
